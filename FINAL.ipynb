```python
#Ex5.14 แบต ตอรี่ชนิดหนึ่งมีอายใช้งาน ฉลี่ย 3 ปี และส่วน บียง บนมาตรฐาน 0.5
#ปี สมมติว่าอายุใช้งานของแบต ตอรี่ชนิดนี้มีการแจกแจงปกติ จงหาความน่าจะ ป็นที่
#แบต ตอรี่ชนิดนี้จะมีอายน้อยกว่า 2.3 ปี

#กรณีการแจกแจงปกติ
mu = 1000 #อายใช้งาน ฉลี่ย 3 ปี 
sigma = 0.5 #ส่วน บียง บนมาตรฐาน 0.5 ปี 
x = 2.3 #มีอายน้อยกว่า 2.3 ปี

prob = st.norm.cdf(x,mu,sigma)
print('P(X<%.2f)=%.4f'%(x,prob))

#กรณีการแจกแจงปกติมาตรฐาน

Z = (x-mu)/sigma

prob=st.norm.cdf(Z)
print('P(Z<%.2f)=%.4f'%(Z,prob))
```

```python
#ข. กรณีไม่ทราบความแปรปรวนของ 2 ประชากร แต่ทราบว่ามีค่าไม่ ท่ากัน และสุ่ม
    #ตัวอย่างนิสิตสาขา CS มา 9 คน มีค่าใช้จ่ายโดย ฉลี่ยต่อ ดือน 6,150 บาท ส่วน
    #บี่ยง บนมาตรฐาน 180 บาท และสุ่มตัวอย่างนิสิตสาขาอื่นๆ มา 11คน มี
    #ค่าใช้จ่าย ฉลี่ยต่อ ดือน 6,070 บาท ส่วน บียง บนมาตรฐาน 200 บาท จงหาช่วง
    #ความ ชื่อมั่นที่ 95% ของผลต่างของค่าใช้จ่ายต่อ ดือน ฉลี่ยของนิสิตสาขา CS
    #และสาขาอื่นๆ
#7.12 ข

d = 0 #ค่าเฉลี่ย = 0
x1 = 6150 #ค่าใช้จ่ายโดย ฉลี่ยต่อ ดือน 6,150 บาท 
x2 = 6070 #ค่าใช้จ่าย ฉลี่ยต่อ ดือน 6,070 บาท
s2_1 = 180**2 #ค่าใช้จ่ายโดย ฉลี่ยต่อ ดือน 6,150 บาท ส่วนบี่ยง บนมาตรฐาน 180 บาท 
s2_2 = 200**2 # ค่าใช้จ่าย ฉลี่ยต่อ ดือน 6,070 บาท ส่วน บียง บนมาตรฐาน 200 บาท 
n1 = 9 #สุ่มตัวอย่างนิสิตสาขา CS มา 9 คน
n2 = 11 #สุ่มตัวอย่างนิสิตสาขาอื่นๆ มา 11คน
alpha = 0.05 #ระดับนัยสำคัญ 95 % เหลือ 5 % = 0.05

#กรณีที่ 3 แจกแจงใดๆ ไม่ทราบค่าความแปรปรวน != T , V
t = ((x1 - x2) - d) / math.sqrt((s2_1 / n1) + (s2_2 / n2))
print('t =', t)

v = (((s2_1 / n1) + (s2_2 / n2)) **2) / (((s2_1 / n1)**2) / (n1 - 1) + (s2_2 / n2) ** 2 / (n2 - 1)) 
print('df =', v)

cv = st.t.ppf([1 - alpha], v) #ค่าวิกฤต
print('cv =', cv)

#ด้านมาก
if t <= cv:
    print('Accept H0') #บริเวรยอมรับ
else:
    print('Reject H0') #บริเวณปฏิเศษ
```

```python
#7.12 ก
#7.12 ตัวอย่าง 7.12 จงทดสอบสมมติฐานที่ว่า ค่าใช้จ่ายโดย ฉลี่ยต่อ ดือนของนิสิตสาขา CS
#มมส. จะสูงกว่าค่าใช้จ่าย ฉลี่ยของนิสิตสาขาอื่นๆ ภายในคณะวิทยาการสารสน ทศ มมส.
#ที่ระดับนัยสำคัญ 0.01 หรือไม่ ถ้า
    #ก. สุ่มตัวอย่างนิสิตสาขา CS มา 60 คน มีค่าใช้จ่ายโดย ฉลี่ยต่อ ดือน 6,150 บาท
    #ส่วน บี่ยง บนมาตรฐาน 180 บาท และสุ่มตัวอย่างนิสิตสาขาอื่นๆ มา 100 คน มี
    #ค่าใช้จ่าย ฉลี่ยต่อ ดือน 6,070 บาท ส่วน บียง บนมาตรฐาน 200 บาท
d = 0 #ค่าเฉลี่ย = 0
x1 = 6150 #ค่าใช้จ่ายโดย ฉลี่ยต่อ ดือน 6,150 บาท 
x2 = 6070 #ค่าใช้จ่าย ฉลี่ยต่อ ดือน 6,070 บาท
s2_1 = 180**2 #ค่าใช้จ่ายโดย ฉลี่ยต่อ ดือน 6,150 บาท ส่วนบี่ยง บนมาตรฐาน 180 บาท 
s2_2 = 200**2 # ค่าใช้จ่าย ฉลี่ยต่อ ดือน 6,070 บาท ส่วน บียง บนมาตรฐาน 200 บาท 
n1 = 60 #สุ่มตัวอย่างนิสิตสาขา CS มา 60 คน
n2 = 100 #สุ่มตัวอย่างนิสิตสาขาอื่นๆ มา 100 คน
alpha = 0.01 #ระดับนัยสำคัญ

#กรณีที่ 2 แจกแจงใดๆ ไม่ทราบค่าความแปรปรวน Z
z =  ((x1-x2) - d)/math.sqrt((s2_1/n1)+(s2_2/n2)) 
print('z =',z)

cv = st.norm.ppf([1-alpha]) #ค่าวิกฤต
print('cv =',cv)

#ด้านมาก
if z<=cv:
    print('Accept H0') #บริเวรยอมรับ
else:
    print('Reject H0') #บริเวณปฏิเศษ
```
