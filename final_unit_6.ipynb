```python
#Ex 6.3 ตัวอย่าง 6.3 จากการสุ่มตัวอย่างขนาด 36 จากประชากรที่มี u/= 78 และมีความแปรปรวน
#a2 = 25 จงหาความน่าจะ ป็นที่ X- จะมากกว่า 79

import scipy.stats as st
import math

n = 36 #ขนาดประชากร
xbar = 79 #ความน่าจะเป็นที x- จะมากกว่า 79 
mu = 78 #ค่าเฉลี่ย
sigma = math.sqrt(25)#ค่าความแปรปรวน

Z = (xbar - mu)/(sigma/math.sqrt(n))# ทฤษฏีที่ 1

prob = 1 - st.norm.cdf(Z) #norm.cdf คือการแจกแจง
print('P(X>%d)=P(Z>%.2f)=%.4f'%(xbar,Z,prob))
```

```python
#Ex 6.4 อายุการใช้งานของโทรศัพท์มือถือ ครื่องหนึ่งมีการแจกแจงปกติ โดยมีอายุ
#การใช้งาน ฉลี่ย 700 ชม. มีส่วน บียง บนมาตรฐาน 40 ชม. ถ้าสุ่มตัวอย่างโทรศัพท์มือถือ
#นี้มา 16 ครื่อง จงหาความน่าจะ ปันที่ตัวอย่างสุ่มของโทรศัพท์มือถือ 16 ครื่องนี้มีอายุใช้

#Ex 6.4 a น้อยกว่า 685
n = 16 #ตัวอย่างสุ่มของโทรศัพท์มือถือ 16 ครื่อง
xbar = 685 #น้อยกว่า 685
mu = 700 #การใช้งาน ฉลี่ย 700 ชม
sigma = 40 #ส่วน บียง บนมาตรฐาน 40 ชม

Z = (xbar - mu)/(sigma/math.sqrt(n)) # ทฤษฏีที่ 1

prob = st.norm.cdf(Z)#norm.cdf คือการแจกแจง
print('P(X<%d)=P(Z<%.2f)=%.4f'%(xbar,Z,prob))
```

```python
#Ex 6.4 b ระหว่าง 700 & 720

n = 16 #norm.cdf คือการแจกแจง
xbar1 = 700 #ระหว่าง 700 & 720
xbar2 = 720 #ระหว่าง 700 & 720
mu = 700 ##การใช้งาน ฉลี่ย 700 ชม
sigma = 40 #ส่วน บียง บนมาตรฐาน 40 ชม

Z1 = (xbar1-mu)/(sigma/math.sqrt(n)) # ทฤษฏีที่ 1
Z2 = (xbar2-mu)/(sigma/math.sqrt(n)) # ทฤษฏีที่ 1

p1 = st.norm.cdf(Z1) #norm.cdf คือการแจกแจง
p2 = st.norm.cdf(Z2) #norm.cdf คือการแจกแจง
print('P(Z2<%.2f)=%.4f, P(Z1<%.2f)=%.4f'%(Z2,p2,Z1,p1))

prob = p2 - p1 #ระหว่างนำมา ลบ กัน
print('P(%d<X<%d)=P(%.2f<Z<%.2f)=%.4f'%(xbar1,xbar2,Z1,Z2,prob))
```

```python
#Ex 6.5 ถ้านิสิตในมหาวิทยาลัยมหาสารคามมีความสูง ฉลี่ย 165.5 ซม. ทำการสุ่ม
#ตัวอย่างนิสิตในมหาวิทยาลัยมหาสารคามมา 30 คน แล้วนำมาวัดส่วนสูง พบว่ามีส่วน
#บียง บนมาตรฐาน ท่ากับ 5.5 ซม. จงหาความน่าจะ ป็นที่นิสิตในมหาวิทยาลัยแห่งนี้มี
#ความสูง ฉลี่ยระหว่าง 164 ถึง 168.5 ซม.

n = 30 #ตัวอย่างนิสิตในมหาวิทยาลัยมหาสารคามมา 30 คน 
xbar1 = 164 #ความสูง ฉลี่ยระหว่าง 164 ถึง 168.5 ซม.
xbar2 = 168.5#ความสูง ฉลี่ยระหว่าง 164 ถึง 168.5 ซม.
mu = 165.5 #ความสูง ฉลี่ย 165.5 ซม.
sigma = 5.5 #บียง บนมาตรฐาน ท่ากับ 5.5 ซม.

Z1 = (xbar1-mu)/(sigma/math.sqrt(n)) # ทฤษฏีที่ 1
Z2 = (xbar2-mu)/(sigma/math.sqrt(n)) # ทฤษฏีที่ 1

p1 = st.norm.cdf(Z1) #norm.cdf คือการแจกแจง
p2 = st.norm.cdf(Z2) #norm.cdf คือการแจกแจง
print(p2, p1)

prob = p2 - p1#ระหว่างนำมา ลบ กัน
print('P(%d<X<%d)=P(%.2f<Z<%.2f)=%.4f'%(xbar1,xbar2,Z1,Z2,prob))
```

```python
#Ex 6.7 IQ ของนิสิต มมส. มีการแจกแจงแบบปกติมีค่า ฉลี่ย 130 ถ้าสุ่มนิสิตมา 25
#คน พบว่า  IQ มีค่า ยง บนมาตรฐาน 5 จงหาความน่าจะ ป็นที่นิสิตจะมี I/ ฉลี่ยมากกว่า 131.059

n = 25 #สุ่มนิสิตมา 25 คน
xbar = 131.059 #ค่าเฉลี่ยมากกว่า  131.059
mu = 130 #ปกติมีค่า ฉลี่ย 130 
s = 5 #ค่าเบี่ยงเบนมาตรฐาน 5 
df = n - 1 ##การแจกแจงองศสอิสระ

T = (xbar-mu)/(s/math.sqrt(n)) #จากสูตร

prob = 1-st.t.cdf(T,df)
print('P(X>%.3f) = P(T(%d)>%.4f) = %.4f'%(xbar,df,T,prob))
```

```python
#Ex 6.8 บริษัทผลิตอาหารกระป้องแห่งหนึ่ง ทราบว่า น้ำหนักบรรจุสุทธิของอาหาร มี
#การแจกแจงแบบปกติ มีความแปรปรวน 5 (กรัม)2ถ้าสุ่มอาหารกระป้องมา 20 กระป้อง จง
#หาความน่าจะ ปันที่ความแปรปรวนของน้ำหนักบรรจุสุทธิของอาหาร 20 กระป้องนี้ มีค่า
#มากกว่า 7 (กรัม)2

n = 20 #สุ่มอาหารกระป้องมา 20 กระป้อง 
s2 =7 #มีค่ามากกว่า 7 (กรัม)2
var = 5 #ความแปรปรวน 5 (กรัม)
df = n-1 #การแจกแจงองศสอิสระ

chi = (df*s2)/var # 19 * 7 / 5 = 26.6000 8 8nv การแจกแจงแบบไคกำลัง 2

p1 = st.chi2.cdf(chi,df) 
print('P1 = %.4f'%p1)

prob =1 - p1
print('P(s2>%d) = P(Chi2(%d)>%.4f) = %.4f'%(s2,df,chi,prob))
```

```python
#Ex 6.9 ถ้ารายได้ของร้านขายอาหารในมหาวิทยาลัยแห่งหนึ่งมีการแจกแจงแบบ
#ปกติ มีส่วน บียง บนมาตรฐาน 25 บาท สุมตัวอย่างร้านขายอาหารมาจำนวน 10 ร้าน
#สอบถามรายได้ต่อวัน จงหาความน่าจะ ปีนที่ความแปรปรวนนี้จะอยู่ระหว่าง 289.50 ถึง
#1019.75 บาท

n = 10 #สุมตัวอย่างร้านขายอาหารมาจำนวน 10 ร้าน
s2_1 = 289.5 #ความแปรปรวนนี้จะอยู่ระหว่าง 289.50 ถึง 1019.75 บาท
s2_2 = 1019.75 #ความแปรปรวนนี้จะอยู่ระหว่าง 289.50 ถึง 1019.75 บาท
var = 25**2 #ส่วน บียง บนมาตรฐาน 25 บาท ดังนั้น ความแปรปรวน a2 = 625

df = n-1 #การแจกแจงองศสอิสระ
chi2_1 = (df*s2_1)/var #การแจกแจงแบบไคกำลัง 2
chi2_2 = (df*s2_2)/var #การแจกแจงแบบไคกำลัง 2

p1 = st.chi2.cdf(chi2_1,df)
p2 = st.chi2.cdf(chi2_2,df)
print('P2 = %.4f , P1 = %.4f'%(p2,p1))

prob = p2 - p1#หาค่าระหว่าง
print('P(%.2f<s2<%.2f) = P(%.4f<Chi2(%d)<%.4f) = %.4f'%(s2_1,s2_2,chi2_1,df,chi2_2,prob))
```

```python
#Ex 6.10 ประชาชนที่อยู่ในจังหวัดหนึ่งมีสัดส่วนของการ ปีนหนี้ ปีน 0.6 และมีการ
#แจกแจงทวินามซึ่งมาจากการสุ่มตัวอย่างประชาชนจำนวน 100 คน จงหาความน่าจะ ป็นที่
#ประชาชนจะ ป็นหนี้อยู่ระหว่าง 50 ถึง 70 คน

n = 100 #สุ่มตัวอย่างประชาชนจำนวน 100 คน
x1 = 50 #ประชาชนจะ ป็นหนี้อยู่ระหว่าง 50 ถึง 70 คน
x2 = 70 #ประชาชนจะ ป็นหนี้อยู่ระหว่าง 50 ถึง 70 คน
p = 0.6 #มีสัดส่วนของการ ปีนหนี้ ปีน 0.6

p_hat1 = x1/n #จากทฤษฏี
p_hat2 = x2/n #จากทฤษฏี

Z1 = (p_hat1-p)/math.sqrt((p*(1-p))/n) #การสูตร
Z2 = (p_hat2-p)/math.sqrt((p*(1-p))/n) #การสูตร

prob1 = st.norm.cdf(Z1) #norm.cdf คือการแจกแจง
prob2 = st.norm.cdf(Z2) #norm.cdf คือการแจกแจง
print('prob2 = %.4f , prob1 = %.4f'%(prob2,prob1)) 

prob=prob2-prob1
print('P(%.2f<p_hat<%.2f) = P(%.4f<Z<%.4f) = %.4f'%(p_hat1,p_hat2,Z1,Z2,prob))
```

```python
#Ex 6.12 ตัวอย่าง 6.11 ผลการสอบวิชาสถิติของนิสิตสองกลุ่มมีการแจกแจงปกติ ด้วยค่า ฉลี่ยและ
#ความแปรปรวนคือ M1 = 50, 0 = 5 และ 42 = 40, 02 = 4 ถ้าสุ่มตัวอย่างนักศึกษามาก
#ลุ่มละ 10 คน จงหาความน่าจะ ป็นที่ผลต่างค่า ลี่ยมีค่ามากกว่า 8 คะแนน

n1 = 10 #สุ่มตัวอย่างนักศึกษามากลุ่มละ 10 คน 
n2 = 10 #สุ่มตัวอย่างนักศึกษามากลุ่มละ 10 คน 
d = 8 #ค่า ลี่ยมีค่ามากกว่า 8 คะแนน
mu1 = 50 #ค่า ฉลี่ย
mu2 = 40  #ค่า ฉลี่ย
var1 = 5 #ความแปรปรวน
var2 = 4 #ความแปรปรวน

Z = (d-(mu1-mu2))/math.sqrt((var1/n1)+(var2/n2)) #นิยามที่ทราบความแปรปรวน

prob = 1 - st.norm.cdf(Z)
print('P(d>%d )= P(Z>%.4f) = %.4f'%(d,Z,prob))
```
