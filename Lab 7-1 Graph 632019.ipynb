```python
import matplotlib.pyplot as plt
profit = np.array([20.5, 18.25, 19.7, 25.2, 26.4, 23.13])
plt.plot(profit)
year = np.arange(1998,2004)
plt.figure(2)
plt.plot(year, profit)
cost = np.array([15.5, 13.25, 14.7, 20.2, 22.4, 20.13])
plt.figure(3)
plt.plot(year, profit,"rx:",year,cost,'bo-')
plt.xlabel('Year')
plt.ylabel('Profit & cost (Million Baht)')
plt.title('Present graph')
plt.show()
```

```python
plt.figure(4)
plt.bar(year,profit)
plt.title('Bar chart1')
plt.xlabel('Year')
plt.ylabel('Profit (Million Baht)')
```

```python
plt.bar(year,profit,color=['y','r','g','b','k','m'])
plt.title('Bar Chart')
plt.xlabel('Year')
plt.ylabel('Profit (Million Baht)')
plt.show()
```

```python
# Data of plot
plt.figure(6)
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0.3, 0)    # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
            autopct='%1.2f%%')

plt.axis('equal')
plt.show()
```
