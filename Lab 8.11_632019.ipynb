```python
#8.11
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

index = np.arange(4)
barwidth = 0.35
a = [1, 2, 3, 4]
b = [5, 4, 2, 1]
plt.figure(1,figsize=[10,10])
plt.subplot(221)
plt.bar(index,a,barwidth,color='r',label='a')
plt.bar(index+barwidth,b,barwidth,color='b',label='b')
plt.xlabel('Index')
plt.ylabel('Million baht')
plt.xticks(index + (barwidth/2),index)
plt.title('a and b data')
plt.legend()

plt.subplot(222)
plt.plot(a,'r--',label='a')
plt.plot(b, 'b',label='b')
plt.xlabel(' Index')
plt.ylabel('Million baht')
plt.xticks(index , index)
plt.title('a and b data')
plt.legend()

plt.subplot(223)
plt.bar(index,a,color='r',label='a')
plt.plot(b, 'b',label='b')
plt.xlabel('Index')
plt.ylabel('Million baht')
plt.xticks(index ,index)
plt.title('a and b data')
plt. legend()
plt.show()
```
