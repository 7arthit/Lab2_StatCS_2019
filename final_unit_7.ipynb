```python
import scipy.stats as st
import math

#Ex 7.5  บริษัทผลิต Projector ได้กล่าวว่า Projector มีอายุการใช้งาน ฉลี่ย 35000
#ชั่วโมง จากการสุ่มตัวอย่าง Projector 10 ครื่อง พบว่ามีอายุการใช้งาน ฉลี่ย 29500
#ชั่วโมง หากทราบว่าอายุการใช้งานมีการแจกแจงปกติด้วยค่าส่วน บียง บนมาตรฐาน
#5000 ชั่วโมง จงทดสอบสมมติฐานว่าคำกล่าวอ้างของบริษัท ปันจริงหรือไม่ ที่ระดับนัยสำคัญ 0.05

#H0 : mu = 35000 
#H1 : mu != 35000
#alpha = 0.05, two - tail

mu = 35000 #อายุการใช้งาน ฉลี่ย 35000 ชั่วโมง
sigma = 5000 #ส่วน บียง บนมาตรฐาน 5000 ชั่วโมง
xbar = 29500 #การใช้งาน ฉลี่ย 29500 ชั่วโมง จาก 10 เครื่อง
n = 10 #สุ่มตัวอย่าง Projector 10 ครื่อง
alpha = 0.05 #ระดับนัยสำคัญ 0.05

# Get the observed value of test statistic.
z = (xbar-mu)/(sigma/math.sqrt(n)) #ทฤษฏีที่ 1 แจกแจงปกติ ทราบค่าความแปรปรวน
print('z =',z)

p = [alpha/2,1-alpha/2] #บริเวณวิกฤต
print(p)

cv = st.norm.ppf(p)#ค่าวิกฤต
print('cv=',cv)

# two tailed test
if z>=cv[0] and z<=cv[1]:
    print( 'Accept H0') #บริเวรยอมรับ
else:
    print('Reject H0') #บริเวณปฏิเศษ
```

```python
# Ex 7.6  จากใบสมัค รียนของมหาวิทยาลัยแห่งหนึ่งกล่าวว่า พื่อประสิทธิภาพของ
#การยนในสถาบัน ทางมหาวิทยาลัยได้จัดชั้น ยนขนาด ล็กโดยแต่ละห้องมีจำนวน
#นักศึกษาไม่ กิน 35คน สุมตัวอย่างห้อง รียนจำนวน 64 ห้อง พบว่ามีจำนวนนักศึกษา รียน
#โดย ฉลี่ย 37 คนและส่วน บียง บนมาตรฐาน 6 คน ท่านคิดว่าในใบสมัค รียนที่กล่าวว่า
#พื่อประสิทธิภาพของการ รียนนั้น ปันจริงหรือไม่ที่ระดับนัยสำคัญ 0.05

# H0 : mu<=35, H1 : mu>35, alpha = 0.05, one-way ด้านมาก
mu = 35 #จัดชั้น ยนขนาด ล็กโดยแต่ละห้องมีจำนวนนักศึกษาไม่ กิน 35คน
s = 6 #ส่วน บียง บนมาตรฐาน 6 คน 
xbar = 37 #มีจำนวนนักศึกษา รียนโดย ฉลี่ย 37 คน
n = 64 #สุมตัวอย่างห้อง รียนจำนวน 64 ห้อง
alpha = 0.05 #ระดับนัยสำคัญ 0.05

# Get the observed value of test statistic.
z = (xbar - mu)/(s/math.sqrt(n)) # ทฤษฏีที่ 2 แจกแจงใดๆ ไม่ทราบค่าความแปรปรวน
print('z =',z)

p = [1-alpha ] #บริเวณวิกฤต
print(p)

cv = st.norm.ppf(p)# st.norm.ppf([1-alpha]) #ค่าวิกฤต
print('cv=',cv)

#one tailed test ด้านมาก
if z <= cv : #z<=Cv[0]
    print('Accept H0') #บริเวรยอมรับ
else:
    print('Reject H0') #บริเวณปฏิเศษ
```

```python
# Ex 7.7  ผู้จัดการคิดว่ากรลดน้ำหนักของลูกค้าจะต้องลดลงอย่างน้อย 45 ปอนด์
#ภายใน 6 ดือน สุมตัวอย่างลูกค้าที่ใช้บริการการลดน้ำหนัก 28 คน น้ำหนักโดย ฉลี่ย
#ลดลง 35 ปอนด์ ส่วนบียง บนมาตรฐาน 20 ปอนด์ ที่ระดับนัยสำคัญ 0.01 จงทดสอบว่า
#ความ ชื่อของผู้จัดการ กินความ ปีนจริงหรือไม่ ถ้านำหนักที่ลดของลูกค้ามีการแจกแจงแบบปกติ

mu = 45 #ลดลงอย่างน้อย 45 ปอนด์ ภายใน 6 ดือน
s = 20 #ส่วนบียง บนมาตรฐาน 20 ปอนด์ 
xbar = 35 #น้ำหนักโดย ฉลี่ยลดลง 35 ปอนด์
n = 28 #สุมตัวอย่างลูกค้าที่ใช้บริการการลดน้ำหนัก 28 คน
alpha = 0.01 #ระดับนัยสำคัญ 0.01
df = n - 1  #การแจกแจงองศาอิสระ

t = (xbar - mu)/(s/math.sqrt(n)) # ทฤษฏีที่ 3 แจกแจงปกติ ไม่ทราบค่าความแปรปรวน
print('t =',t)

p = [alpha] #บริเวณวิกฤต
print(p)

cv = st.t.ppf(p,df) #ค่าวิกฤต
print('cv =',cv)

#ด้านน้อย
if t >= cv:
    print('Accept H0') #บริเวรยอมรับ
else:
    print('Reject H0') #บริเวณปฏิเศษ
```

```python
# Ex 7.8 ชื่อว่าร้อยละ 70% ของผู้ชมที่ ป็นสตรี ห็นด้วยกับกรรมการตัดสินการ
#ประกวดนางงาม ที่ให้สาวงามจากประ ทศไทยได้รับรางวัลชนะ ลิศประ ภทแต่งกายประจำ
#ชาติ ในปีที่ประ ทศไทย ป็น จ้าภาพการประกวดนางงามจักรวาลปี 2005 จากการสุ่ม
#ตัวอย่างสตรีจำนวน 500 คน มีคน หันด้วย 300 คน จงทดสอบสมมติฐานดังกล่าวว่า ป็น
#จริงหรือไม่ ที่ระดับนัยสำคัญ 0.05

#H0 : p = 0.7
#H1 : p! = 0.7

p = 0.7 #ร้อยละ 70% ของผู้ชมที่ ป็นสตรี 
x = 300 #มีคน หันด้วย 300 คน
n = 500 #จากการสุ่มตัวอย่างสตรีจำนวน 500 
p_hat = x/n # จากการสุ่มตัวอย่างสตรีจำนวน 500 คน มีคน หันด้วย 300 คน
alpha = 0.05 #

z = (p_hat - p)/math.sqrt((p*(1-p))/n) #จากสูตร
print('z =',z)

cv = st.norm.ppf([alpha/2,1-alpha/2]) #ค่าวิกฤต
print('cv =',cv)

if z >=cv[0] and z<=cv[1]:
    print('Accept H0') #บริเวรยอมรับ
else:
    print('Reject H0') #บริเวณปฏิเศษ
```

```python
#7.9 ผู่จัดการต้องการทราบว่าสัดส่วนของผู้บริโภคแชมพูห้อลอริอัล อย่างมาก
#ป็น 35 % หรือไม่ จึงทำการสุมตัวอย่างลูกค้าจำนวน 100 คน มีผู้ใช้แชมพูยี่ห้อนี้ 40 คน
#จงทดสอบสมมติฐานดังกล่าว ปันจริงหรือไม่ ที่ระดับนัยสำคัญ 0.01

#H0 : p <= 0.7
#H1 : p > 0.7

p = 0.35 # อย่างมาก ป็น 35 % 
x = 40 #มีผู้ใช้แชมพูยี่ห้อนี้ 40 คน
n = 100 #การสุมตัวอย่างลูกค้าจำนวน 100 คน
p_hat = x/n #การสุมตัวอย่างลูกค้าจำนวน 100 คน มีผู้ใช้แชมพูยี่ห้อนี้ 40 คน
alpha = 0.01 #ระดับนัยสำคัญ 

z = (p_hat - p)/math.sqrt((p*(1-p))/n) #จากสูตร
print('z =',z)

cv = st.norm.ppf([1-alpha/2]) #ค่าวิกฤต
print('cv =',cv)

#ด้านมาก
if z<=cv:
    print('Accept H0') #บริเวรยอมรับ
else:
    print('Reject H0') #บริเวณปฏิเศษ
```

```python
#7.10 วิศวกรต้องการพิสูจน์ว่า ครื่องจักรที่ทำการบรรจน้ำอัดลมกระป้องชนิด
#หนึ่งมื่อทำการบรจุน้ำอัดลมกระป้องแล้ว ให้ค่าความแปรปรวน กินกว่า 0.01 มิลลิลิตร2
#หรือไม่ จึงได้สุ่ม ก็บตัวอย่างน้ำอัดลมกระป้องที่ทำการบรรจุ สร็จแล้วมาจำนวน 20
#กระป้อง พบว่ากลุ่มตัวอย่างดังกล่าวมีค่าความแปรปรวน ท่ากับ 0.0153 มิลลิลิตร อยาก
#ทราบว่า ครื่องจักรที่ทำการบรรจุน้ำอัดลมกระป้อง ครื่องดังกล่าวจะต้องได้รับการแก้ไข
#หรือไม่ ถ้าหากมีข้อกำหนดว่า จะต้องแก้ไข ครื่องจักรหากค่าความแปรปรวนมากกว่า
#0.01 มิลลิลิตร2 มื่อกำหนดระดับนัยสำคัญ 0.05 และประชากรของปริมาตรน้ำอัดลม
#กระป้อง ปี๊นแบบ Normal distribution

#H0 : var = 0.01
#H1 : var > 0.01

var = 0.01 #ให้ค่าความแปรปรวน กินกว่า 0.01 มิลลิลิตร2หรือไม่ 
s2 = 0.0153 #พบว่ากลุ่มตัวอย่างดังกล่าวมีค่าความแปรปรวน ท่ากับ 0.0153 มิลลิลิตร
n = 20 #สุ่ม ก็บตัวอย่างน้ำอัดลมกระป้องที่ทำการบรรจุ สร็จแล้วมาจำนวน 20 กระป้อง 
alpha = 0.05 #ระดับนัยสำคัญ 0.05 
df = n-1 #การแจกแจงองศาอิสระ

#chi2 = x-
chi2 = (df*s2)/var #จากสุตร
print('chi2 =',chi2)

cv = st.chi2.ppf([1-alpha],df) #ค่าวิกฤต
print('cv =',cv)

#ด้านมาก
if chi2 <= cv:
    print('Accept H0') #บริเวรยอมรับ
else:
    print('Reject H0') #บริเวณปฏิเศษ
```

```python
#7.11 ตัวอย่าง 7.11โรงงานแห่งหนึ่ง ต้องการสั่งซื้อชิ้นส่วนจำนวนหนึ่ง พื่อนำมาผลิตสินค้า ที่
#มีความแปรปรวนของความยาวไม่ กิน 0.0001 มิลลิ มตร2 ผู้จัดการต้องการทดสอบว่า
#ชิ้นส่วนที่สั่งมามีคุณภาพตามความต้องการหรือไม่ จึงสุ่มตัวอย่างชิ้นส่วนมา 25 ชิ้น ได้
#ความแปรปรวนได้ 0.0002 มิลลิ มตร2 อยากทราบว่าผู้จัดการจะยอมรับชิ้นส่วนจำนวนนั้น
#หรือไม่ที่ระดับนัยสำคัญ 0.01

#H0 : var <= 0.0001
#H1 : var > 0.0001

var = 0.0001 #ต้องการสั่งซื้อชิ้นส่วนจำนวนหนึ่ง ที่มีความแปรปรวนของความยาวไม่ กิน 0.0001 มิลลิ มตร2 
s2 = 0.0002 #ความแปรปรวนได้ 0.0002 มิลลิ มตร2
n = 25 #สุ่มตัวอย่างชิ้นส่วนมา 25 ชิ้น
alpha = 0.01 # ระดับนัยสำคัญ
df = n-1

chi2 = (df*s2)/var #จากสูตร
print('chi2 =',chi2)

cv = st.chi2.ppf([1-alpha],df)  #ค่าวิกฤต
print('cv =',cv)

#ด้านมาก
if chi2 <= cv:
    print('Accept H0') #บริเวรยอมรับ
else:
    print('Reject H0') #บริเวณปฏิเศษ
```

```python
#7.12 ตัวอย่าง 7.12 จงทดสอบสมมติฐานที่ว่า ค่าใช้จ่ายโดย ฉลี่ยต่อ ดือนของนิสิตสาขา CS
#มมส. จะสูงกว่าค่าใช้จ่าย ฉลี่ยของนิสิตสาขาอื่นๆ ภายในคณะวิทยาการสารสน ทศ มมส.
#ที่ระดับนัยสำคัญ 0.01 หรือไม่ ถ้า
    #ก. สุ่มตัวอย่างนิสิตสาขา CS มา 60 คน มีค่าใช้จ่ายโดย ฉลี่ยต่อ ดือน 6,150 บาท
    #ส่วน บี่ยง บนมาตรฐาน 180 บาท และสุ่มตัวอย่างนิสิตสาขาอื่นๆ มา 100 คน มี
    #ค่าใช้จ่าย ฉลี่ยต่อ ดือน 6,070 บาท ส่วน บียง บนมาตรฐาน 200 บาท
#7.12 ก

d = 0 #ค่าเฉลี่ย = 0
x1 = 6150 #ค่าใช้จ่ายโดย ฉลี่ยต่อ ดือน 6,150 บาท 
x2 = 6070 #ค่าใช้จ่าย ฉลี่ยต่อ ดือน 6,070 บาท
s2_1 = 180**2 #ค่าใช้จ่ายโดย ฉลี่ยต่อ ดือน 6,150 บาท ส่วนบี่ยง บนมาตรฐาน 180 บาท 
s2_2 = 200**2 # ค่าใช้จ่าย ฉลี่ยต่อ ดือน 6,070 บาท ส่วน บียง บนมาตรฐาน 200 บาท 
n1 = 60 #สุ่มตัวอย่างนิสิตสาขา CS มา 60 คน
n2 = 100 #สุ่มตัวอย่างนิสิตสาขาอื่นๆ มา 100 คน
alpha = 0.01 #ระดับนัยสำคัญ

#กรณีที่ 2 แจกแจงใดๆ ไม่ทราบค่าความแปรปรวน Z
z =  ((x1-x2) - d)/math.sqrt((s2_1/n1)+(s2_2/n2)) 
print('z =',z)

cv = st.norm.ppf([1-alpha]) #ค่าวิกฤต
print('cv =',cv)

#ด้านมาก
if z<=cv:
    print('Accept H0') #บริเวรยอมรับ
else:
    print('Reject H0') #บริเวณปฏิเศษ
```

```python
#ข. กรณีไม่ทราบความแปรปรวนของ 2 ประชากร แต่ทราบว่ามีค่าไม่ ท่ากัน และสุ่ม
    #ตัวอย่างนิสิตสาขา CS มา 9 คน มีค่าใช้จ่ายโดย ฉลี่ยต่อ ดือน 6,150 บาท ส่วน
    #บี่ยง บนมาตรฐาน 180 บาท และสุ่มตัวอย่างนิสิตสาขาอื่นๆ มา 11คน มี
    #ค่าใช้จ่าย ฉลี่ยต่อ ดือน 6,070 บาท ส่วน บียง บนมาตรฐาน 200 บาท จงหาช่วง
    #ความ ชื่อมั่นที่ 95% ของผลต่างของค่าใช้จ่ายต่อ ดือน ฉลี่ยของนิสิตสาขา CS
    #และสาขาอื่นๆ
#7.12 ข

d = 0 #ค่าเฉลี่ย = 0
x1 = 6150 #ค่าใช้จ่ายโดย ฉลี่ยต่อ ดือน 6,150 บาท 
x2 = 6070 #ค่าใช้จ่าย ฉลี่ยต่อ ดือน 6,070 บาท
s2_1 = 180**2 #ค่าใช้จ่ายโดย ฉลี่ยต่อ ดือน 6,150 บาท ส่วนบี่ยง บนมาตรฐาน 180 บาท 
s2_2 = 200**2 # ค่าใช้จ่าย ฉลี่ยต่อ ดือน 6,070 บาท ส่วน บียง บนมาตรฐาน 200 บาท 
n1 = 9 #สุ่มตัวอย่างนิสิตสาขา CS มา 9 คน
n2 = 11 #สุ่มตัวอย่างนิสิตสาขาอื่นๆ มา 11คน
alpha = 0.05 #ระดับนัยสำคัญ 95 % เหลือ 5 % = 0.05

#กรณีที่ 3 แจกแจงใดๆ ไม่ทราบค่าความแปรปรวน != T , V
t = ((x1 - x2) - d) / math.sqrt((s2_1 / n1) + (s2_2 / n2))
print('t =', t)

v = (((s2_1 / n1) + (s2_2 / n2)) **2) / (((s2_1 / n1)**2) / (n1 - 1) + (s2_2 / n2) ** 2 / (n2 - 1)) 
print('df =', v)

cv = st.t.ppf([1 - alpha], v) #ค่าวิกฤต
print('cv =', cv)

#ด้านมาก
if t <= cv:
    print('Accept H0') #บริเวรยอมรับ
else:
    print('Reject H0') #บริเวณปฏิเศษ
```

```python
#Exem 1 จากใบสมัครเรียนของมหาวิทยาลัยแห่งหนึ่งกล่าวว่าเพื่อประลิทธิภาพของการเรียนในสถาบัน ทางมหาวิทยาลัยได้จัดชั้น
#เรียนขนาดเล็กโดยแต่ละห้องมีจำนวนนักศึกษาเท่ากับ 37 คน และส่วนเบียงเบนมาตรฐาน 7 คน สมตัวอย่างห้องเรียน
#จำนวน 20 ห้อง พบว่ามีจำนวนนักศึกษาเรียนโดยเฉลี่ย 37 น ท่านคิดว่าในใบสมัครเริยนที่กล่าวว่าเพื่อประสิทธิภาพ
#ของการเรียนนั้นเป็นจริงหรือไม่ที่ระดับนัยสำคัญ 0.01

mu = 37 #แต่ละห้องมีจำนวนนักศึกษาเท่ากับ 37 คน
sigma = 7# ส่วนเบียงเบนมาตรฐาน 7 คน
xbar = 37 #พบว่ามีจำนวนนักศึกษาเรียนโดยเฉลี่ย 37 น
n = 20 #สมตัวอย่างห้องเรียนจำนวน 20 ห้อง
alpha = 0.01 #ระดับนัยสำคัญ 0.01

z = (xbar - mu) / (sigma / math.sqrt(n)) # ทฤษฏีที่ 1 แจกแจงปกติ ทราบค่าความแปรปรวน
print('z =', z)
 
p = [alpha / 2, 1 - alpha / 2] #บริเวณวิกฤต
print(p)

cv = st.norm.ppf(p) )#ค่าวิกฤต
print('cv=', cv)

# two tailed test
if z >= cv[0] and z <= cv[1]:
    print('Accept H0') #บริเวรยอมรับ
else:
    print('Reject H0') #บริเวณปฏิเศษ
```

```python
#Exem 2 บริษัทผลิตยางรถยนต์ยี่ห้อหนึ่ง โฆษณาว่า ยางรถยนต์ยี่ห้อนี้สามารถวิ่งได้ระยะทางเฉลี่ยไม่น้อยกว่า 70,000 ก.ม.
#นักวิจัยสงสัยว่า บริษัทโมษณาเกินจริง เชื่อว่ายางรถยนต์ยี่ห้อนี้จะวิ่งได้ระยะทางเฉลี่ยจริง ๆ น้อยกว่า 70,000 ก.ม. ลุ่ม
#ยางรถยนต์มา 16 เส้น พบว่ายางรถยนต์ยี่ห้อนี้สามารถวิ่งได้ระยะทางเฉลียที่ 65,000 และมีส่วนเบียงเบนมาตรฐาน
#10,000 ก.ม. ที่ช่วงความเชื่อมั่น 90%

mu = 70000 #วิ่งได้ระยะทางเฉลี่ยจริง ๆ น้อยกว่า 70,000 ก.ม
s = 10000 #ส่วนเบียงเบนมาตรฐาน 10,000 ก.ม.
xbar =  65000 #สามารถวิ่งได้ระยะทางเฉลียที่ 65,000
n = 16 # สุ่มยางรถยนต์มา 16 เส้น
alpha = 0.1 # ระดับนัยสำคัญ 10 %

z = (xbar - mu) / (s / math.sqrt(n)) # ทฤษฏีที่ 2 แจกแจงใดๆ ไม่ทราบค่าความแปรปรวน
print('z =', z)

p = [1 - alpha] #บริเวณวิกฤต
print(p)

cv = st.norm.ppf(p) #ค่าวิกฤต
print('cv=', cv) 

#ด้านน้อย
if z >= cv:
    print('Accept H0') #บริเวรยอมรับ
else:
    print('Reject H0') #บริเวณปฏิเศษ
```

```python
#Exem 11 กระบวนการผลิตในโรงงานหนึ่ง จะผลิตสินค้า สีย 20% ฝ่ายผลิตจึงได้ดำ นินการ
#ปรับปรุงกระบวนการผลิตใหม่ แล้วสุ่มตัวอย่างสินค้ามา 100 ชิ้น พบว่า ป็นสินค้า 
#สีย 16 ชิ้น จะ ชื่อได้หรือไม่ว่า การปรับปรุงกระบวนการผลิตจะทำให้สินค้า สีย
#ลดลง ที่ระดับนัยสำคัญ 0.01

p = 0.2 # 20 %
x = 16 #พบว่า ป็นสินค้า สีย 16 ชิ้น
n = 100 #สุ่มตัวอย่างสินค้ามา 100 ชิ้น
p_hat = x / n #สุ่มตัวอย่างสินค้ามา 100 ชิ้น พบว่า ป็นสินค้า สีย 16 ชิ้น
alpha = 0.01 # ระดับนัยสำคัญ 0.01

z = (p_hat - p) / math.sqrt((p * (1 - p)) / n) #การสูตร บท 6
print('z =', z)

cv = st.norm.ppf([alpha / 2, 1 - alpha / 2]) #ค่าวิกฤต
print('cv =', cv)

if cv[0] <= z <= cv[1]:
    print('Accept H0') #บริเวรยอมรับ
else:
    print('Reject H0') #บริเวณปฏิเศษ
```

```python
#Exem 17 จากประสบการณ์ของอาจารย์ท่านหนึ่ง ได้ระบว่า วลาที่นักศึกษาใช้ในการทดสอบ
#ย่อยวิชาสถิดิ บื้องต้น มีการแจกแจงแบบปกติ มีส่วน บียง บนมาตรฐานน้อยกว่า 6
#นาที จึงสุมนักศึกษามา 20 คน หาค่า บี่ยง บนมาตรฐานได้ 4.51 อยากทราบว่าสิ่ง
#ที่อาจารย์ท่านนี้กล่าวอ้าง ปีนจริงหรือไม่ ที่ระดับนัยสำคัญ 0.05

var = 6 #มีการแจกแจงแบบปกติ มีส่วน บียง บนมาตรฐานน้อยกว่า 6 นาที 
s2= 4.51#ค่า บี่ยง บนมาตรฐานได้ 4.51 
n = 20 #สุมนักศึกษามา 20 คน
alpha = 0.05 #ระดับนัยสำคัญ 0.05
df = n - 1

chi2 = (df * s2) / var #จากสุตร 7.7.3
print('chi2 =', chi2)

cv = st.chi2.ppf([1 - alpha], df) #ค่าวิกฤต
print('cv =', cv)

#ด้านน้อย
if chi2 >= cv:
    print('Accept H0') #บริเวรยอมรับ
else:
    print('Reject H0') #บริเวณปฏิเศษ
```
